```python
#importation des librairies
import pandas as pd
```

```python
### Lecture de fichiers ###
# Taxo
taxo = pd.read_excel('2023.03.13_R&I Taxonomy (2).xlsx', sheet_name='R&I Taxonomy')
#sdds
sdds_taxo = pd.read_excel('taxo_sdds_FGI.xlsx')
```

```python
taxo
```

```python
### Mise en forme du fichier de taxonomie ###
# Garder les éléments de la famille "Formula"
formula = taxo.loc[taxo['Family'] == 'Formula']
# Garder les éléments de la SF "MD General Information"
general_info_taxo = formula.loc[formula['Sub-family'] == 'MD Formula General Information']
# Sélection des éléments non obsolètes
general_info_taxo_val = general_info_taxo.loc[general_info_taxo['Taxonomy maturity (attribute level)'].isin(['Validated','On going','To be validated'])]
# Calcul du nombre de BOAs non nuls
count_boa_taxo = general_info_taxo_val['Attribute with Trigram'].count()
```

```python
general_info_taxo_val
```

```python
### Mise en forme du fichier sdds
sdds_taxo.columns = sdds_taxo.iloc[0]
```

```python
sdds_taxo
```

```python
## jointure entre les tables
joined_tab = pd.merge(general_info_taxo_val,sdds_taxo,left_on='Attribute with Trigram',right_on='Attribute')
joined_tab
```

```python
### Table des différences
merged_table = pd.merge(general_info_taxo_val,sdds_taxo,left_on='Attribute with Trigram',right_on='Attribute', how='outer', indicator=True)

# sélectionner les enregistrements uniques
unique_records = merged_table.loc[merged_table['_merge'] != 'both']
```

```python
unique_records
```

```python
### Téléchargement des excels
joined_tab.to_excel('sdds_taxo_FGI_join.xlsx')
unique_records.to_excel('sdds_taxo_FGI_unique_rec.xlsx')
```
